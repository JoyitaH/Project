```python
def Category(x):
    if(x == "0=Blood Donor"):
        return "0"
    elif(x == "0s=suspect Blood Donor"):
        return "1"
    elif(x == "1=Hepatitis"):
        return "2"
    elif(x == "2=Fibrosis"):
        return "3"
    elif(x == "3=Cirrhosis"):
        return "4"
```

```python
df["Category"] = df["Category"].apply(lambda x : Category(x))
df
```

```python
def Sex(y):
    if(y == "f"):
        return "0"
    elif(y == "m"):
        return "1"
```

```python
df["Sex"] = df["Sex"].apply(lambda y : Sex(y))
df
```

Calculating VIF 
Ideally VIF should be less than or equal to 5

```python
# Import Library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    #Calculating VIF
    vif=pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return(vif)
```

```python
X = df.iloc[:,:-1]
calc_vif(X)
```

```python
#spliting independent features and target label
X = df.iloc[:,1:]  #independent features
Y = df["Category"]  #dependent feature or target 
```

```python
#SKLearn for train and test split

from sklearn.model_selection import train_test_split
from sklearn.metrics import *

#train ==> i will teach my model, frame mathematical equation
#test
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8, random_state = 0)
```

```python
Y_train
```

```python
from sklearn.preprocessing import MinMaxScaler

min_scalar = MinMaxScaler()
min_scalar.fit_transform(X_train) 
min_scalar.transform(X_test) 
```

```python
#Model Development and Prediction

#sklearn model to load Logistic regression package
from sklearn.linear_model import LogisticRegression
#max_iter=1200000, dual=False , class_weight="balanced", penalty="none"
logreg = LogisticRegression(solver = "sag", max_iter=10000)

#fit training data
logreg.fit(X_train, Y_train)

#predict unseen data on trained model
Y_pred = logreg.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
conf_matrix = confusion_matrix(Y_test, Y_pred)
```

```python
conf_matrix
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
accuracy
```

```python
from sklearn.metrics import precision_score
precision = precision_score(Y_test, Y_pred, average='weighted')
precision
```

```python
from sklearn.metrics import recall_score
recall = recall_score(Y_test, Y_pred, average='weighted')
recall
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

Y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(Y_test,  Y_pred, pos_label=None)
auc = roc_auc_score(Y_test, Y_pred, average='weighted')
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.xlabel("False Positives")
plt.ylabel("True Positives")
plt.legend(loc=4)
plt.show()
```
