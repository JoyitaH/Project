```python
def age(x):
    if(x > 15 and x <= 30):
        return "Youth"
    elif(x > 30 and x <= 45):
        return "Adult"
    elif(x > 45 and x <= 60):
        return "Middle Aged"
    else:
        return "Old Age"
```

```python
df["Age_Groups"] = df["age"].apply(lambda x: age(x))
df
```

```python
def bmi(y):
    if(y > 18.5 and y < 25):
        return "Ideal"
    elif(y <= 18.5):
        return "Underweight"
    elif(y >= 25):
        return "Overweight"
```

```python
df["BMI_Range"] = df["bmi"].apply(lambda y : bmi(y))
df
```

```python
#To add the new columns to the parent Data Set in the CSV File
df.to_csv("C:/Users/Joyita/Data_Science/Data/test_insurance.csv")
```
